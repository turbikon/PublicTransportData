## Public Transport Monitoring System
### First Stage

It is complicate developing the planned whole system from the beginning, so it is better to split it in stages.

The first stage covers the easiest possible approche for the system, that include the next:
- Cam system
    - for get-in and get-out events
- GPS system
- Accelerometer system
- Payment system
    - for cash payment
    - for fixed trip cost
- Alert system, 
    - for get-in and get-out events
    - for accelerometer measures
    - for GPS signals

The excluded systems are listed here just to be clear about the remaining work.
- Cam system
    - for face-recognition 
    - for passenger-distribution
- Payment systems
    - for card payment
    - for mobile payment
    - for variable trip cost
- Communication systems
    - for mobile communication
    - for WLAN communication
- Alert system

*The focused subsystems for the first stage can change along progress*
